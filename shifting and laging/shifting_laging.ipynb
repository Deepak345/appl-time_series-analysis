```python
data['prev_day_price'] = data['Price'].shift(1)
data
```

```python
data['1 day change']= data['Price'] - data['prev_day_price']
data
```

```python
data['5 day % return'] = data['Price'] - (data['Price'].shift(5)*100)/data['Price'].shift(5)
data
```
